```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
file = open("Source_data_for_CFR_vaccine_map_abridged.csv", 'r')

data_s = []

for line in file:
    source = line.strip("\n").split(',')
    data_s.append(source)

data = np.zeros(len(data_s), dtype=[('disease', np.str_, 1),
                               ('position',  np.float_, 2),
                               ('magnitude', np.float_, 1)])

for i in range(len(data_s)):
    data['disease'][i] = data_s[i][0]
    data['position'][i] = (float(data_s[i][1]), float(data_s[i][2]))
    data['magnitude'][i] = float(data_s[i][3])
```

```python
fig = plt.figure(figsize=(14,10))
ax = plt.subplot(1,1,1)

earth = Basemap(projection='mill')

earth.drawcoastlines(color='0.50', linewidth=0.25)
earth.fillcontinents(color='0.95')

plt.title("Measles Outbreak - 100568134", fontsize =18)

max_mag = max(data['magnitude'][:])
print(max_mag)

plt.scatter(data['position'][:,0],
            data['position'][:,1], (data['magnitude'] / max_mag),
            color = 'red')
plt.show()

# points do not show up, no point in continuing
```
