```python
# Michael Stergianis (100568134)
from matplotlib import pyplot as plt
from matplotlib import patches as pch

import calendar
import numpy as np
%matplotlib inline
```

```python
# load files from cwd, make sublist containing only data files
import os
files = os.listdir()
temperature_files = []
for file in files:
    if "temperature_2012.dat" in file:
        temperature_files.append(file)
```

```python
# add files to the plot, in this case we know we only have three
plt.figure(num=None, figsize=(10, 7), dpi=80, facecolor='w', edgecolor='k')
plt.plotfile(temperature_files[0], color='red', linewidth='2.5', newfig=False, names=calendar.month_abbr[1:13])
plt.plotfile(temperature_files[1], color='blue', linewidth='2.5', newfig=False, names=calendar.month_abbr[1:13])
plt.plotfile(temperature_files[2], color='green', linewidth='2.5', newfig=False, names=calendar.month_abbr[1:13])

plt.xticks(np.arange(0,12), calendar.month_abbr[1:13])
line1 = pch.Patch(color='red')
line2 = pch.Patch(color='blue')
line3 = pch.Patch(color='green')
plt.legend((line1, line2, line3), ('Montreal', 'Toronto', 'Vancouver'), loc='upper right' )
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.title("2012 Average Temperatures (100568134)", fontsize=14)
# data ends at December, so it shows empty entries but prevents obscuration of data
plt.axis([0,11,-10,25])

plt.show()
```
