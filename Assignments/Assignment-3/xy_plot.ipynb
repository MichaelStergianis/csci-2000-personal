```python
# Michael Stergianis (100568134)
import numpy as np
import matplotlib.pyplot as plt
# from matplotlib import patches as pch
%matplotlib inline
```

```python
points = open("xy_points.dat", 'r')

x_list = []
y_list = []
# read our floating point entries into x_list and y_list
for entry in points.readline().split(' '):
    x_list.append(float(entry))

for entry in points.readline().split(' '):
    y_list.append(float(entry))
# make arrays from list
x = np.array(x_list)
y = np.array(y_list)
# there are more memory efficient ways to do the above,
# but we only have 1000 elements and it is very readable this way

# logical indexing to remove elements that we do not want
# square
x_sub0 = x[ (x < -2) ]
y_sub0 = y[ (x < -2) ]
x_sub1 = x_sub0[ (y_sub0 > -5) & (y_sub0 < 0) ]
y_sub1 = y_sub0[ (y_sub0 > -5) & (y_sub0 < 0) ]
# circle
x_rad = x[ ( ( ( (x)**2 + (y-5)**2 )**(1/2) ) <= 5 ) ]
y_rad = y[ ( ( ( (x)**2 + (y-5)**2 )**(1/2) ) <= 5 ) ]

plt.figure(num=None, figsize=(8, 8), dpi=80, facecolor='w', edgecolor='k')
plt.scatter(x, y, color='r')
plt.scatter(x_sub1, y_sub1, color = 'b')
plt.scatter(x_rad, y_rad, color='g')
plt.axis([-10,10,-10,10])
plt.title("Hilighted Points (100568134)", fontsize=14)

plt.show()
```
