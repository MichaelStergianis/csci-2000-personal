```python
# Michael Stergianis (100568134)
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline
```

```python
countries = "USA", "CHN", "GBR", "RUS", "KOR", "GER", "FRA", "ITA", "HUN", "AUS", "OTHER"
golds = [46, 38, 29, 24, 13, 11, 11, 8, 8, 7, 107]
```

```python
# get colormap
cs = cm.Set2(np.arange(1, 255, 23))
# plot our pie chart on a figure that is square
plt.figure(num=None, figsize=(8, 8), dpi=80, facecolor='w', edgecolor='k')
plt.title("2012 Olympic Gold Medals (100568134)", fontsize=15)
plt.pie(golds, explode=[.2,.2,.2,.2,.2,.2,.2,.2,.2,.2,.2], labels=countries, 
        colors=cs, autopct='%1.1f%%', shadow=True, radius=1)

plt.show()
```
